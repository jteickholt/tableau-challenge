### This notebook was used to bring in the individual month of 2014, choose a random sample of 5% for each month, then combine them to provide a sampe of the data for 2014. This was then output to a .csv that was used in Tableau

```python
#store the data we get as a dataframe
import pandas as pd

#mathematical operations on lists
import numpy as np

import os
```

### Read in each months data and convert to dataframes.

```python
# define path to files 

jan14_file = os.path.join('.' , 'data' , 'jan14_bikedata.csv')
feb14_file = os.path.join('.' , 'data' , 'feb14_bikedata.csv')
mar14_file = os.path.join('.' , 'data' , 'mar14_bikedata.csv')
apr14_file = os.path.join('.' , 'data' , 'apr14_bikedata.csv')
may14_file = os.path.join('.' , 'data' , 'may14_bikedata.csv')
jun14_file = os.path.join('.' , 'data' , 'jun14_bikedata.csv')
jul14_file = os.path.join('.' , 'data' , 'jul14_bikedata.csv')
aug14_file = os.path.join('.' , 'data' , 'aug14_bikedata.csv')
sep14_file = os.path.join('.' , 'data' , 'sep14_bikedata.csv')
oct14_file = os.path.join('.' , 'data' , 'oct14_bikedata.csv')
nov14_file = os.path.join('.' , 'data' , 'nov14_bikedata.csv')
dec14_file = os.path.join('.' , 'data' , 'dec14_bikedata.csv')

```

```python
# read in the files

jan14 = pd.read_csv(jan14_file)
feb14 = pd.read_csv(feb14_file)
mar14 = pd.read_csv(mar14_file)
apr14 = pd.read_csv(apr14_file)
may14 = pd.read_csv(may14_file)
jun14 = pd.read_csv(jun14_file)
jul14 = pd.read_csv(jul14_file)
aug14 = pd.read_csv(aug14_file)
sep14 = pd.read_csv(sep14_file)
oct14 = pd.read_csv(oct14_file)
nov14 = pd.read_csv(nov14_file)
dec14 = pd.read_csv(dec14_file)
```

```python
# convert to dataframes

jan14_df = pd.DataFrame(jan14)
feb14_df = pd.DataFrame(feb14)
mar14_df = pd.DataFrame(mar14)
apr14_df = pd.DataFrame(apr14)
may14_df = pd.DataFrame(may14)
jun14_df = pd.DataFrame(jun14)
jul14_df = pd.DataFrame(jul14)
aug14_df = pd.DataFrame(aug14)
sep14_df = pd.DataFrame(sep14)
oct14_df = pd.DataFrame(oct14)
nov14_df = pd.DataFrame(nov14)
dec14_df = pd.DataFrame(dec14)
```

### The months are now combined into a single dataframe.  Also, did some checks of  the final data

```python
# combine all months

frames = [jan14_samp_df, feb14_samp_df, mar14_samp_df, apr14_samp_df, may14_samp_df, jun14_samp_df, jul14_samp_df, aug14_samp_df, 
          sep14_samp_df, oct14_samp_df, nov14_samp_df, dec14_samp_df]

# convert to a dataframe

all2014_df = pd.concat(frames)

# all2014_df[["starttime", "stoptime"]] = all2014_df[["starttime", "stoptime"]].apply(pd.to_datetime,format='%m/%d/%Y %H:%M:%S')

# print out some details of how may records

print(all2014_df.shape)

print()

# check for null values

print(all2014_df.isnull().sum())

print()

# check to see how many have unknown for gender

print(all2014_df['gender'].value_counts())
                                                                                        
all2014_df.head(5)
                                                                                        
                                                                                    
```
